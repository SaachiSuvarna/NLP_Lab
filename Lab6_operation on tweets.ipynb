# CLean the twitter data

#The function below taks the list of words or text as input and returns a cleaner set of words.

```python
#Data cleaning
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
lem=WordNetLemmatizer()
sw=stopwords.words('english')
words=df['text'].to_list()
words="".join(i.lower() for i in words)
words=re.sub('[^\w\s]','',words).split()
words=[lem.lemmatize(i) for i in words if i not in sw]
words
```

#15 most frequently occuring Bigrams and grigrams occur the most in the sample of tweets

```python
import matplotlib.pyplot as plt
def graph(k):
    ngram=nltk.ngrams(words,k)
    sr=pd.Series(ngram).value_counts()
    print(sr[:15])
    sr[:15].plot.bar()
    plt.title('15 Most frequency occuring for k='+str(k))
    plt.xlabel('n-gram')
    plt.ylabel('No. of Occurances')
    plt.show()
```

```python
graph(2)
graph(3)
```
